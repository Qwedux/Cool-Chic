```python
import re


def _extract_row(filename: str) -> str:
    match = re.search(
        r"job_(\d+)_img_rate_([0-9]*\.?[0-9]+)\.pth$",
        filename,
    )
    if not match:
        return ""
    job_idx, rate = match.groups()
    # return f"{kodim} {job_idx} {rate}"
    return f"{job_idx} {rate}"


def extract_table(raw_ls_output: str) -> str:
    entries = raw_ls_output.split()
    rows = [_extract_row(entry) for entry in entries]
    rows = [row for row in rows if row]
    rows.sort(key=lambda x: int(x.split()[0]))
    return "\n".join(rows)


print(extract_table(two_by_two_0))
```
