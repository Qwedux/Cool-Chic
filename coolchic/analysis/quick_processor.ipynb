```python
import re


def _extract_row(filename: str) -> str:
    match = re.search(
        r"kodim(\d+)_img_rate_([0-9]*\.?[0-9]+)\.pth$",
        filename,
    )
    if not match:
        print("cannot parse match")
        return ""
    job_idx, rate = match.groups()
    # return f"{kodim} {job_idx} {rate}"
    return f"{int(job_idx)} {rate}"


def extract_table(raw_ls_output: str) -> str:
    entries = raw_ls_output.split()
    rows = [_extract_row(entry) for entry in entries]
    rows = [row for row in rows if row]
    rows.sort(key=lambda x: int(x.split()[0]))
    return "\n".join(rows)

def select_second(s):
    lines = s.split("\n")
    first_lines = lines[:4]
    rest = lines[4:]
    return first_lines + rest[::2], rest[1::2]

def tuple_fmt(l: list[str]) -> str:
    return "\n".join(l)

tmp = """
2026_01_28__17_31_05__trained_coolchic_kodak_kodim02_img_rate_3.0625627040863037.pth
2026_01_28__17_32_04__trained_coolchic_kodak_kodim01_img_rate_3.660256862640381.pth
2026_01_28__17_59_55__trained_coolchic_kodak_kodim03_img_rate_2.9735944271087646.pth
2026_01_28__18_02_36__trained_coolchic_kodak_kodim04_img_rate_3.114288091659546.pth
2026_01_29__00_18_23__trained_coolchic_kodak_kodim05_img_rate_3.8054745197296143.pth
2026_01_29__00_23_45__trained_coolchic_kodak_kodim06_img_rate_3.271472215652466.pth
2026_01_29__01_48_42__trained_coolchic_kodak_kodim07_img_rate_3.0202174186706543.pth
2026_01_29__02_06_05__trained_coolchic_kodak_kodim08_img_rate_3.9350883960723877.pth
2026_01_29__07_04_56__trained_coolchic_kodak_kodim09_img_rate_3.022674560546875.pth
2026_01_29__07_58_31__trained_coolchic_kodak_kodim10_img_rate_3.1478164196014404.pth
2026_01_29__09_33_24__trained_coolchic_kodak_kodim11_img_rate_3.258312702178955.pth
2026_01_29__10_28_13__trained_coolchic_kodak_kodim12_img_rate_2.9295308589935303.pth
2026_01_29__13_49_06__trained_coolchic_kodak_kodim13_img_rate_4.056516170501709.pth
2026_01_29__14_42_26__trained_coolchic_kodak_kodim14_img_rate_3.5328173637390137.pth
2026_01_29__17_10_53__trained_coolchic_kodak_kodim15_img_rate_3.0701630115509033.pth
2026_01_29__17_10_58__trained_coolchic_kodak_kodim16_img_rate_3.09297513961792.pth
2026_01_29__21_20_26__trained_coolchic_kodak_kodim17_img_rate_3.251532554626465.pth
2026_01_29__21_45_26__trained_coolchic_kodak_kodim18_img_rate_3.8350977897644043.pth
2026_01_29__23_54_36__trained_coolchic_kodak_kodim19_img_rate_3.2642297744750977.pth
2026_01_30__00_57_38__trained_coolchic_kodak_kodim20_img_rate_2.82743763923645.pth
2026_01_30__04_01_59__trained_coolchic_kodak_kodim21_img_rate_3.2762324810028076.pth
2026_01_30__05_45_34__trained_coolchic_kodak_kodim22_img_rate_3.672572612762451.pth
2026_01_30__06_41_46__trained_coolchic_kodak_kodim23_img_rate_3.11757230758667.pth
2026_01_30__08_47_53__trained_coolchic_kodak_kodim24_img_rate_3.5807414054870605.pth
"""
print(extract_table(tmp))
```
