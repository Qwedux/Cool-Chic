```python
from dataclasses import dataclass, field

@dataclass
class A:
    x1: tuple[int] = (5,)
    x2:int = field(default_factory=lambda: 4)

n1 = A()
n2 = A()
n1.x1 = (10, )
n2.x2 = 20

print(f"n1: {n1}")
print(f"n2: {n2}")
print(f"The two fields have equal ids for x: {id(n1.x1) == id(n2.x1)}")
print(f"The two fields have equal ids for x2: {id(n1.x2) == id(n2.x2)}")
```
